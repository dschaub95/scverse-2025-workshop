```python
%load_ext autoreload
%autoreload 2

from claude_agent_sdk import query, ClaudeAgentOptions
import shutil
import glob
import os

while not os.path.exists("pyproject.toml"):
    os.chdir("..")
```

```python
working_dir = "working_dir"


async def main():
    options = ClaudeAgentOptions(
        system_prompt="You are an expert Python developer",
        permission_mode="acceptEdits",
        cwd=working_dir,
        model="claude-haiku-4-5",
        mcp_servers={"anndata-mcp": {"command": "uvx", "args": ["anndata-mcp"]}},
        setting_sources=["project"],
        allowed_tools=["mcp__anndata-mcp"],
    )

    async for message in query(
        prompt="Compile a short report.md file that summarizes the .h5ad object in the data folder using the anndata-mcp tools",
        options=options,
    ):
        print(message)


os.makedirs(f"{working_dir}/data", exist_ok=True)

for h5ad_file in glob.glob("data/*.h5ad"):
    shutil.copy(h5ad_file, f"{working_dir}/data")


await main()
```
